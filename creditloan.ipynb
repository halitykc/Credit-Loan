```python
testdf = trainData.fillna({
    "Loan_Amount_Term":342,             # NaN değerleri Ortalaması ile değiştirdik
    "LoanAmount":146.41216216216216,    # NaN değerleri Ortalaması ile değiştirdik
    "Credit_History": 1 })              #verilerin buyuk coğunlugunun kredi geçmişi olduğu için NaN değerleride veri geçmişi var olarak işaretledim
testdf =testdf.dropna(subset=["Self_Employed","Gender","Dependents"])  # "Self_Employed","Dependents" ve "Gender" da ki NaN değerleri kesitiremediğim için NaN olan satırları sildim
```

```python
for col in testdf.columns:
    print(f"\"{col}\" Sütununda : \"{testdf[col].isna().sum()}\" adet eksik veri mevcut.")
```

```python
testdf.count()
```

```python
trainData = testdf.drop(columns=["Loan_ID"],axis=1)
trainData.head()
```

```python
for col in trainData.columns:
    print(trainData[col].unique())
```

```python
XTrain = pd.get_dummies(trainData, columns=["Dependents","Property_Area"],dtype=int)
```

```python
XTrain.head()
```

```python
XTrain["Gender"] = XTrain["Gender"].replace("Male",0).replace("Female",1)
XTrain["Married"] = XTrain["Married"].replace("No",0).replace("Yes",1)
XTrain["Education"] = XTrain["Education"].replace("Not Graduate",0).replace("Graduate",1)
XTrain["Self_Employed"] = XTrain["Self_Employed"].replace("No",0).replace("Yes",1)
XTrain["Loan_Status"] = XTrain["Loan_Status"].replace("N",0).replace("Y",1)
```

```python
XTrain.head()
```

```python
y = XTrain["Loan_Status"].values
X = XTrain.drop("Loan_Status",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python

Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
rfc = RandomForestClassifier()
rfc.fit(Xtrain,ytrain)
```

```python
tahminler = rfc.predict(Xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
conf = confusion_matrix(ytest,tahminler)
acc = accuracy_score(ytest,tahminler)
cr = classification_report(ytest,tahminler)
```

```python
import seaborn as sns
print("Confusion Matrix \n",conf)
print("Accuracy:", acc*100,"%")
print(cr)
```
